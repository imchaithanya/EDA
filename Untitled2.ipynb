 Step 1  Think of data as people standing on a number line

Imagine you have numbers that represent ages or salaries.
Each person stands on a line according to their value:

<--- small ----|---- middle ----|---- large --->


Now, we want one number to summarize where most people stand  the center.

Step 2  The Mean and Median behave differently
 The mean cares about distance

It adds everyones value together and divides by the count.
So if one person stands very far out, they pull the mean in that direction.
Mean is like a balance point  even one heavy weight far on one side tilts it.

 The median cares about position

It only looks at whos in the middle when everyone stands in a line.
So even if one person stands far away, the median doesnt move  it only checks order, not distance.

| Shape            | Tail direction                | Mean vs Median | What happens          | Which to use |
| ---------------- | ----------------------------- | -------------- | --------------------- | ------------ |
| **Symmetrical**  | None                          | Mean  Median  | Both represent center | **Mean**    |
| **Right-skewed** | Right tail (few big numbers)  | Mean > Median  | Mean pulled up        | **Median**  |
| **Left-skewed**  | Left tail (few small numbers) | Mean < Median  | Mean pulled down      | **Median**  |


Mean moves. Median resists.
Whenever data has a tail (skewed), the mean follows it  the median stays true to the center.
Thats why you use mean for symmetrical data and median for skewed data.

| Step                 | Question                             | Why it matters                             |
| -------------------- | ------------------------------------ | ------------------------------------------ |
| **1 Check %**      | How much data is missing?          | If >40%, you might drop the column         |
| **2 Check Center** | Whats the mean, median, and mode? | This tells you how the data is distributed |
| **3 Check Curve**  | How is the data shaped (skewness)? | This decides *which* central value to use  |


| Type of Column           | Shape / Skew | Best Strategy        | Why                                |
| ------------------------ | ------------ | -------------------- | ---------------------------------- |
| **Numeric**              | Symmetrical  | Fill with **mean**   | Mean = median; either works        |
| **Numeric**              | Right-skewed | Fill with **median** | Mean pulled up by high outliers    |
| **Numeric**              | Left-skewed  | Fill with **median** | Mean pulled down by small outliers |
| **Categorical**          | N/A          | Fill with **mode**   | Mode = most frequent category      |
| **Too many NaNs (>60%)** | N/A          | **Drop column**      | Too incomplete for reliable fill   |


| Column       | Missing % | Shape               | Fill with               | Why                                |
| ------------ | --------- | ------------------- | ----------------------- | ---------------------------------- |
| **Age**      | 19.8%     | Right-skewed        | **Median (28)**        | Mean pulled up by older passengers |
| **Fare**     | 0.0%      | Highly right-skewed | **Median (if missing)** | Protect from rich outliers         |
| **Embarked** | 0.2%      | Categorical         | **Mode (S)**         | Most common port                   |
| **Cabin**    | 77%       | Useless             | **Drop**               | Too incomplete                     |

 Step 1  What is EDA (Exploratory Data Analysis)?

EDA is the process of exploring, visualizing, and understanding your dataset before you start building any model.

It answers this question:

What does my data actually look like, and what patterns, relationships, or problems exist inside it?

You dont jump to ML yet  you first make the data talk to you.

| Phase                         | Focus                                   | Example                             |
| ----------------------------- | --------------------------------------- | ----------------------------------- |
| **1 Univariate Analysis**   | Study each column individually          | Histogram of Age, Fare              |
| **2 Bivariate Analysis**    | Study relation between two columns      | Survival vs Sex, Survival vs Pclass |
| **3 Multivariate Analysis** | Study interactions of multiple features | Survival vs Age + Sex + Class       |
